```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import folium

df=pd.read_excel('MTi-670-8A1G6-2021-12-21-12-04.xlsx')
lat=df['LAT']
lon=df['LON']
```

```python
def merge(list1, list2):
      
    merged_list = [(list1[i], list2[i]) for i in range(0, len(list1))]
    return merged_list


latLon=merge(lat,lon)

```

```python
m = folium.Map(location=latLon[0],zoom_start=15)


m
```

```python
folium.PolyLine(latLon,
                color='red',
                weight=7.5,
                opacity=0.8).add_to(m)
   

m

```
